```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
```

```python
matches = pd.read_csv('ipl/matches.csv')
```

```python
plt.hist(matches['season'], rwidth=0.9)
```

```python
sns.countplot(x='season', data=matches)
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='winner', data=matches)
plt.xticks(rotation='vertical')
```

```python
sns.countplot(x='toss_decision', data=matches)
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='season',data=matches,hue='toss_decision')
```

```python
plt.figure(figsize=(10,6))
ax = matches['toss_winner'].value_counts().plot.bar(color=sns.color_palette('cubehelix'),width=0.9)

for p in ax.patches:
    ax.annotate(format(p.get_height()), (p.get_x() + 0.3, p.get_height() + 1))
```
