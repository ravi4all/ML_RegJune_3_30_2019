```python
deliveries = pd.read_csv('ipl/deliveries.csv')
```

```python
deliveries.shape
```

```python
deliveries.head()
```

```python
temp_df = deliveries[deliveries['dismissal_kind'] == 'caught']
```

```python
pd.value_counts(temp_df['fielder']).head(10)
```

```python
temp_df = deliveries.dropna(subset=['dismissal_kind'])
```

```python
data_2 = temp_df[temp_df['dismissal_kind'] != 'run out']
```

```python
pd.value_counts(data_2['bowler']).head(10)
```

```python
data_3 = deliveries.groupby('batsman')['batsman_runs'].sum()
```

```python
data_3.sort_values(ascending=False).head(10)
```

```python
deliveries.groupby(['batsman','match_id'])['batsman_runs'].sum().sort_values(ascending=False).head(10)
```
